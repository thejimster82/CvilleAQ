```python
import matplotlib.pyplot as plt
import plotly.express as px
import numpy as np
import pandas as pd
import requests
import os
import datetime
from bs4 import BeautifulSoup
```

##### Selection of 10 Charlottesville PWS's

```python
pwsIDs = [105, 171, 111, 136, 112, 80, 106, 176, 134, 8]
```

```python
# [Lat (N), Lon (W), Elev (ft)]
pwsAttr = [[38.035,78.492,500],[38.046,78.495,581],[38.050,78.484,463],[38.029,78.477,472],[38.029,78.470,449],[38.030,78.513,np.NaN],[38.019,78.482,453],[38.059,78.505,581],[38.029,78.463,394],[38.030,78.513,np.NaN]]
```

##### List of Dates

```python
dates = []
start =  datetime.date(2019, 1, 1)
end = datetime.date(2019, 12, 31)
current = start
while current <= end:
    dates.append(str(current))
    current += datetime.timedelta(days=1)
```

```python
#A sample to get dataframe format to later append to.
url = "https://www.wunderground.com/dashboard/pws/KVACHARL105/table/2019-01-01/2019-01-01/daily"
html = requests.get(url)    
soup = BeautifulSoup(html.text, 'html')
table = soup.find_all('table')[3]  # The third table on the page is the one we want
df = pd.read_html(str(table))[0]
df["date"] = np.NaN
df["pws"] = np.NaN
df = df[:1]  #only returning the first NaN row to set columns for appending, will later be dropped

#The Algorithm
for date in dates:
    for pws in pwsIDs:
        url = "https://www.wunderground.com/dashboard/pws/KVACHARL" + str(pws) + "/table/" + date + "/" + date + "/daily"
        html = requests.get(url)
        soup = BeautifulSoup(html.text, 'html')
        try:    #if table is not there, there is no data
            table = soup.find_all('table')[3]
            tempDF = pd.read_html(str(table))[0]
            tempDF["date"] = date
            tempDF["pws"] = pws
            df = df.append(tempDF)
            print(date)
        except IndexError:  
            continue       #will skip the day instead of throwing error

#Dropping the first row of each (the ones with all NaN's)    
df = df.dropna(subset=['Time']) 

df.head()
```

```python
df['Temperature'] = df['Temperature'].astype(str).str[:-2]
df['Dew Point'] = df['Dew Point'].astype(str).str[:-2]
df['Humidity'] = df['Humidity'].astype(str).str[:-2]
df['Speed'] = df['Speed'].astype(str).str[:-4]
df['Gust'] = df['Gust'].astype(str).str[:-4]
df['Pressure'] = df['Pressure'].astype(str).str[:-3]
df['Precip. Rate.'] = df['Precip. Rate.'].astype(str).str[:-3]
df['Precip. Accum.'] = df['Precip. Accum.'].astype(str).str[:-3]
df['Solar'] = df['Solar'].astype(str).str[:-5]
```

```python
df['Temperature'] = df['Temperature'].apply(pd.to_numeric, errors='coerce')
df['Dew Point'] = df['Dew Point'].apply(pd.to_numeric, errors='coerce')
df['Humidity'] = df['Humidity'].apply(pd.to_numeric, errors='coerce')
df['Speed'] = df['Speed'].apply(pd.to_numeric, errors='coerce')
df['Gust'] = df['Gust'].apply(pd.to_numeric, errors='coerce')
df['Pressure'] = df['Pressure'].apply(pd.to_numeric, errors='coerce')
df['Precip. Rate.'] = df['Precip. Rate.'].apply(pd.to_numeric, errors='coerce')
df['Precip. Accum.'] = df['Precip. Accum.'].apply(pd.to_numeric, errors='coerce')
df['Solar'] = df['Solar'].apply(pd.to_numeric, errors='coerce')
```

```python
df.to_csv("weather2019.csv", index=False)
```
