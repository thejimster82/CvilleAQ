```python
import pandas as pd
import numpy as np
import pykrige as pk
import pykrige.kriging_tools as kt
from pykrige.ok import OrdinaryKriging
```

```python
# latitude coords from 38.00 to 38.08
# longitute coords from -78.53 to -78.45

data = np.array([[.01998, .0564, 1.707865],
                 [.03562, .0381, 3.800915],
                 [.02758, .01481, 1.611111],
                 [.02557, .01339, 1.977870],
                 [.04745, .04697, 2.125756],
                 [.02576, .0607, 2.442308],
                 [.03061, .05094, 2.935176],
                 [.02988, .04595, 1.951181]])
                  
gridx = np.arange(0.0, 0.8, 0.05)
gridy = np.arange(0.0, 0.8, 0.05)
```

```python
OK = OrdinaryKriging(data[:, 0], data[:, 1], data[:, 2], variogram_model='linear',
                     verbose=False, enable_plotting=False)
```

```python
z, ss = OK.execute('grid', gridx, gridy)
```
